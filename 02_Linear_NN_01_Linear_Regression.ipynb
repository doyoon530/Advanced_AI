# Revisiting Linear Nueral Network

### Obtain the linear line that passes two points (1,3) and (2,5) by using `torch.nn.Linear, optimizer, and MSELoss`

```python
import matplotlib.pyplot as plt
import numpy as np
import torch
from torch import optim
from torch import nn


inputs = torch.FloatTensor([[1.],[2.]])
targets = torch.FloatTensor([[3.], [5.]])

network = nn.Linear(1,1, bias=True)
optimizer = optim.SGD(network.parameters(), lr= 0.01)
loss_criterion = nn.MSELoss()

num_trains = 3000

for i in range(num_trains):
    optimizer.zero_grad()
    y = network(inputs)
    loss = loss_criterion (y, targets)
    loss.backward()
    optimizer.step()
    print(f"[{i}]\t Loss:{loss.item():.5f}\t Eq:y={network.weight.item():.3f}x+{network.bias.item():.3f}")
    
    
x = torch.arange(0, 5, 0.1)
x = x.reshape(x.shape[0], 1)
y = network(x)
y = y.data

plt.figure(figsize=(10,6))
plt.scatter(inputs, targets, s=100)
plt.plot(x,y,'r-')    
```

### What happends for three points `(1, 3), (2, 5), (3, 9)`

```python
inputs = torch.FloatTensor([[1, 2, 3]]).T # [[1,2,3]] == [1,2,3].T
targets = torch.FloatTensor([[3, 5, 9]]).T

network = nn.Linear(1,1, bias=True)
optimizer = optim.SGD(network.parameters(), lr= 0.01)
loss_criterion = nn.MSELoss()

num_trains = 3000

for i in range(num_trains):
    optimizer.zero_grad()
    y = network(inputs)
    loss = loss_criterion (y, targets)
    loss.backward()
    optimizer.step()
    print(f"[{i}]\t Loss:{loss.item():.5f}\t Eq:y={network.weight.item():.3f}x+{network.bias.item():.3f}")

x = torch.arange(0, 5, 0.1)
x = x.reshape(x.shape[0], 1)
y = network(x)
y = y.data

plt.figure(figsize=(10,6))
plt.scatter(inputs, targets, s=100)
plt.plot(x,y,'r-')    
```

# Regression Using Linear Neural Network


```python
import matplotlib.pyplot as plt
import numpy as np
import torch
from torch import nn

inputs = np.array([[3.3], [4.4], [5.5], [6.71], [6.93], [4.168],
                    [9.779], [6.182], [7.59], [2.167], [7.042],
                    [10.791], [5.313], [7.997], [3.1]], dtype=np.float32)

targets = np.array([[1.7], [2.76], [2.09], [3.19], [1.694], [1.573],
                    [3.366], [2.596], [2.53], [1.221], [2.827],
                    [3.465], [1.65], [2.904], [1.3]], dtype=np.float32)

plt_1 = plt.figure(figsize=(10, 6))
plt.scatter(inputs, targets)
plt.show()
```

## Data preparation

```python
inputs = torch.tensor(inputs, dtype=torch.float32)
targets = torch.tensor(targets, dtype=torch.float32)
```

## Design a neural network

`neural network can be defined by class using nn.Module or nn.Linear`

```python
# Linear Regression Model
class linearRegression(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(1, 1)  # input and output is 1 dimension

    def forward(self, x):
        out = self.linear(x)
        return out
model = linearRegression()
```

```python
model = nn.Linear(1,1)
```

## Training

```python
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=1e-4)


num_epochs = 1000
for epoch in range(num_epochs):
    optimizer.zero_grad()
    # forward
    out = model(inputs)
    loss = criterion(out, targets)
    # backward    
    loss.backward()
    optimizer.step()

    if (epoch+1) % 20 == 0:
        print(f'Epoch[{epoch+1}/{num_epochs}], loss: {loss.item():.6f}', end=' ')
        for name, param in model.named_parameters():
            print (name, param.data.numpy(), end=' ')
        print()
model.eval()
with torch.no_grad():
    predict = model(inputs)
predict = predict.data.numpy()

fig = plt.figure(figsize=(10, 5))
plt.plot(inputs.numpy(), targets.numpy(), 'ro', label='Original data')
plt.plot(inputs.numpy(), predict, label='Fitting Line')
plt.legend() 
plt.show()
```

## Evaluation

```python
x = torch.tensor([[1, 2, 3, 4]], dtype=torch.float32).reshape(-1, 1)
model.eval()
out = model(x)
print(out)
#model.train()

fig = plt.figure(figsize=(10, 5))
plt.plot(inputs.numpy(), targets.numpy(), 'ro', label='Original data')
plt.plot(inputs.numpy(), predict, 'b-',label='Fitting Line')
plt.plot(x.detach().numpy(), out.detach().numpy(), 'bo', label='Predicted')
plt.legend() 
plt.show()
```
